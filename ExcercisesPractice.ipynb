```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
company_sales = pd.read_csv('/content/company_sales_data.csv')
```

```
company_sales
```

```
X = company_sales['month_number'].tolist()
Y = company_sales['total_profit'].tolist()
```

```
plt.plot(X,Y)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title("Company Profit per month")
plt.xticks(company_sales['month_number'])
plt.yticks([100000,200000,300000,400000,500000]);
```

```
plt.plot(X,Y, linestyle='--',label='profit data of last year',marker='o',markerfacecolor='k',color='r',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title("Company Sales data per month")
plt.legend(loc='lower right')
plt.xticks(company_sales['month_number'])
plt.yticks([100000,200000,300000,400000,500000]);
```

```
company_sales.columns
```

```
facecream1 = company_sales['facecream'].tolist()
facewash1 = company_sales['facewash'].tolist()
toothpaste1 = company_sales['toothpaste'].tolist()
bathingsoap1 = company_sales['bathingsoap'].tolist()
shampoo1 = company_sales['shampoo'].tolist()
moisturizer1 = company_sales['moisturizer'].tolist()
months = company_sales['month_number'].tolist()
```

```
plt.plot(months,facecream1, linestyle = '-',marker = 'o', markerfacecolor='b',color='b', linewidth=3, label='facecream data')
plt.plot(months,facewash1, linestyle='-', marker='o',markerfacecolor='orange', color='orange', linewidth=3, label='facewash data')
plt.plot(months,toothpaste1, linestyle='-',marker='o', markerfacecolor='green',color='green',linewidth=3,label='tootpaste data')
plt.plot(months,bathingsoap1, linestyle='-', marker='o', markerfacecolor='red',color='red', linewidth=3,label='bathsoap data')
plt.plot(months,shampoo1, linestyle='-',marker='o',markerfacecolor='purple',color='purple',linewidth=3, label='shampoo data')
plt.plot(months,moisturizer1,linestyle='-', marker='o', markerfacecolor='brown',color='brown', linewidth=3, label='moisturerizer data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales Data')
plt.legend(loc='upper left')
plt.xticks(company_sales['month_number'])
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000]);
```

```
plt.scatter(months,toothpaste1, label = 'Toothpaste sales data')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.legend(loc='upper left')
plt.grid(True,linestyle='--', linewidth = 1)
plt.xticks(company_sales['month_number'])
plt.yticks([4500,5000,6000,6500,7000,7500,8000]);
```

```
plt.bar([a-0.25 for a in months], facecream1,label='facewash and facecream data',width=0.25, align='edge')
plt.bar([a+.25 for a in months], facewash1,label='facewash and facecream data', width=-0.25, align= 'edge')
plt.xlabel('Month number')
plt.ylabel('Sales unit in Number')
plt.legend(loc='upper left')
plt.xticks(company_sales['month_number'])
plt.yticks([500,1000,1500,2000,2500,3000,3500])
plt.grid(True, linestyle='--',linewidth=1)
plt.title('Facewash and facecream sales data')
```

```
plt.bar(months,bathingsoap1, width=.75 )
plt.xticks(company_sales['month_number'])
plt.yticks([2000,4000,6000,8000,10000,12000,14000])
plt.xlabel('Month Number')
plt.ylabel('Sales units in numbers')
plt.title('Bath soap data')
plt.grid(True,linestyle='--', linewidth=1)
plt.savefig('C:\Desktop\SDLLearn\PyNativeExcercises\salesdatabathsoap.png', dpi=150)
plt.show()
```

```
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(Y,profit_range);
```

```
labeldata = ['Facecream','facewash', 'toothpaste','bathing soap', 'shampoo', 'moisturizer' ]
sales_data = [company_sales['facecream'].sum(), company_sales['facewash'].sum(),company_sales['toothpaste'].sum(),company_sales['bathingsoap'].sum(),company_sales['shampoo'].sum(),company_sales['moisturizer'].sum()]
```

```
company_sales.columns
```

```
plt.axis('equal')
plt.pie(sales_data, labels=labeldata,autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('sales data')
plt.show()
```

```
fig , axarr = plt.subplots(2, sharex=True)
axarr[0].plot(months,bathingsoap1, label= 'Sales data of bathing soap', linestyle='-',marker='o', markerfacecolor='black',color='black',linewidth=3)
axarr[1].plot(months,facewash1, label='Sale of facewash',linestyle='-',marker='o', markerfacecolor='red',color='red',linewidth=3)

axarr[0].set_title('Sales data of bathingsoap')
axarr[1].set_title('Sales data of facewash')

plt.xlabel('Month Number')
plt.ylabel('Sale units in number')
plt.xticks(months)
plt.show()
```
